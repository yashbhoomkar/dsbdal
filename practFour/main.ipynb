```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.isnull().sum()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
```

```python
df["deck"] = df["deck"].ffill()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
sns.countplot(data = df , x = "sex" , hue = "survived")
```

```python
sns.countplot(data = df , x = "class" , hue = "survived")
```

```python
plt.figure(figsize = (25 , 10))
sns.countplot(data =df , x = "age" , hue = "survived")
plt.show()
```

```python
sns.histplot(data=df, x="fare", bins=30)  # Histogram with KDE (Kernel Density Estimate)
```

```python
sns.boxplot(data = df , y  = "fare" , hue = "survived")
```
