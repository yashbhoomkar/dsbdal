```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("amazon.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["rating_count"] = df["rating_count"].ffill()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.head()
```

```python
df["discount_percentage"] =  df["discount_percentage"].astype("str")
```

```python
df["discount_percentage"] = df["discount_percentage"].str.replace("%" , "")
```

```python
df["discount_percentage"] = df["discount_percentage"].astype("float")
df["discount_percentage"] = df["discount_percentage"] / 100
```

```python
df["discounted_price"] = df["discounted_price"].astype("string")
```

```python
df["discounted_price"] = df["discounted_price"].str.replace("₹" , "")
df["discounted_price"] = df["discounted_price"].str.replace("," , "")
```

```python
df["actual_price"] = df["actual_price"].astype("string")
df["actual_price"] = df["actual_price"].str.replace("₹" , "")
df["actual_price"] = df["actual_price"].str.replace("," , "")
```

```python
df = df.drop(columns = ["product_id"])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors = "coerce")
df["actual_price"] = pd.to_numeric(df["actual_price"], errors = "coerce")
df["rating"] = pd.to_numeric(df["rating"], errors = "coerce")
```

```python
df["rating_count"].astype("string")
df["rating_count"] = df["rating_count"].str.replace("," , "")
df["rating_count"] = pd.to_numeric(df["rating_count"], errors = "coerce")

```

```python
sns.boxplot(data = df)
```

```python
df.dropna(inplace = True)
```
